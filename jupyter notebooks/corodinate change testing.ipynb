```python
import math as math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10,10)) # for the first figure:

def fine_field_cart_rotation(xField, yField, phi):
    '''
    Rotate a the cordinate system about the z axis
    so we can allign our control components to be perpendicular
    and parallel to the optical zero.
    '''
    # do a rotation about the z axis.
    xFieldPrime = xField * math.cos(phi) + yField * math.sin(phi)
    yFieldPrime = yField * math.cos(phi) - xField * math.sin(phi)
    
    return((xFieldPrime, yFieldPrime))
    
    
innital_field = (0, 2) # x and y field

x = []
y = []

for i in range(0, 360):
    # store one datapoint
    tx, ty = fine_field_cart_rotation(innital_field[0], innital_field[1], np.deg2rad(i))
    x.append(tx)
    y.append(ty)

#print(x)
plt.plot(x, y)

```

```python
# let's create a test shape to rotate with our rotate function.
x = np.array(range(100)) # create x values that are evenly spaced
y = -abs(-1*x + 30) # make an absolute value graph (a check mark)


plt.figure(figsize=(11,5)) # set the figsize 
plt.subplot(121) # setup a two plot figure notation is 1 row 2 colums and place this plot in position 1 
plt.plot(x, y) # first make a plot of the absolue value graph
plt.plot(0,0,'o') # plot a dot at the origin so we can better see the rotation of the graph later
plt.xlim([-150,150]) # lock in the limits of the graph so it has the same scaling after rotaion. 
plt.ylim([-150,150])

# now let's rotate it by a few degrees
xRotated = [] # initalize the empty arrays
yRotated = []
angleDeg = 45 # angle in degrees to rotate the coridinate system 

for i, yData in enumerate(y):
    # store one datapoint
    tx, ty = fine_field_cart_rotation(x[i], yData, np.deg2rad(angleDeg)) # fine_field_cart takes radians
    xRotated.append(tx)
    yRotated.append(ty)

#plt.figure(figsize=(5,10))
plt.subplot(122)
plt.plot(xRotated, yRotated)
plt.plot(0,0,'o') # plot a dot at the origin
plt.xlim([-150,150])
plt.ylim([-150,150])
plt.show()

```
