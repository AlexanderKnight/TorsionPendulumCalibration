```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit



%matplotlib inline
data1 = np.genfromtxt('./../data/quadtrantSearchData.txt', delimiter = ' ')

data1 *= 1000000

def linFit(x, m, b):
    return m*x + b

fitParams, fitCovar = curve_fit(linFit, data1[:,0], data1[:,1])
print(fitParams)

linX = np.linspace(min(data1[:,0]), max(data1[:,0]), 1000)


plt.figure(figsize=(10,8))
plt.plot(data1[:,0], data1[:,1], 'o')
plt.plot(linX, linFit(linX, fitParams[0], fitParams[1]))
plt.grid()
plt.show()
```
